# EDA och visualisering\n
\n
En notebook for snabb explorativ dataanalys med Pandas, Matplotlib och Seaborn.

```python
from pathlib import Path\n
\n
import numpy as np\n
import pandas as pd\n
import matplotlib.pyplot as plt\n
import seaborn as sns\n
\n
sns.set_theme(style='whitegrid')
```

```python
raw_dir = Path('../data/raw')\n
processed_dir = Path('../data/processed')\n
processed_dir.mkdir(parents=True, exist_ok=True)
```

```python
# Demo-data som gar att kora direkt\n
rng = np.random.default_rng(42)\n
df = pd.DataFrame({\n
    'category': rng.choice(['A', 'B', 'C'], size=300, p=[0.4, 0.35, 0.25]),\n
    'value': rng.normal(loc=50, scale=12, size=300).round(2),\n
    'age': rng.integers(18, 70, size=300),\n
})\n
df.head()
```

```python
# Datakontroll\n
display(df.shape)\n
display(df.isna().sum())\n
display(df.describe(include='all'))
```

```python
# Distribution\n
plt.figure(figsize=(8, 4))\n
sns.histplot(df['value'], bins=25, kde=True)\n
plt.title('Distribution av value')\n
plt.xlabel('value')\n
plt.tight_layout()
```

```python
# Jamforelse mellan kategorier\n
plt.figure(figsize=(8, 4))\n
sns.boxplot(data=df, x='category', y='value')\n
plt.title('Value per kategori')\n
plt.tight_layout()
```

```python
# Enkel feature engineering\n
df['age_group'] = pd.cut(df['age'], bins=[17, 29, 44, 59, 100], labels=['18-29', '30-44', '45-59', '60+'])\n
summary = (\n
    df.groupby(['category', 'age_group'], observed=False)['value']\n
      .mean()\n
      .reset_index()\n
      .rename(columns={'value': 'avg_value'})\n
)\n
summary.head()
```

```python
# Exportera sammanstallning\n
out_path = processed_dir / 'eda_summary.csv'\n
summary.to_csv(out_path, index=False)\n
out_path
```
