Use this notebook as a starting point for loading, exploring, and exporting tabular data.

```python
from pathlib import Path\n
import pandas as pd
```

```python
# Paths\n
raw_dir = Path('../data/raw')\n
processed_dir = Path('../data/processed')\n
processed_dir.mkdir(parents=True, exist_ok=True)\n
\n
# Example file path (replace with your own file)\n
csv_path = raw_dir / 'input.csv'\n
csv_path
```

```python
# Load data (uncomment when input file exists)\n
# df = pd.read_csv(csv_path)\n
\n
# Demo dataframe so the notebook runs out of the box\n
df = pd.DataFrame({\n
    'name': ['Anna', 'Bo', 'Carla'],\n
    'score': [85, 92, 78],\n
    'passed': [True, True, False],\n
})\n
\n
df.head()
```

```python
# Basic profiling\n
display(df.info())\n
display(df.describe(include='all'))
```

```python
# Simple transformation\n
summary = (\n
    df.groupby('passed', as_index=False)['score']\n
      .mean()\n
      .rename(columns={'score': 'avg_score'})\n
)\n
summary
```

```python
# Export\n
out_path = processed_dir / 'summary.csv'\n
summary.to_csv(out_path, index=False)\n
out_path
```
