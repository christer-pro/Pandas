## Vad är Pandas?

**Pandas** är ett bibliotek (ett färdigt verktyg) i Python som gör det enkelt att arbeta med **tabelldata** –  
alltså data som ser ut som en Excel-tabell, med rader och kolumner.

Tänk dig att du har en lista med elever och deras betyg:

| Namn   | Betyg | Klass |
|--------|-------|-------|
| Anna   | 85    | 9A    |
| Bo     | 92    | 9B    |
| Carla  | 78    | 9A    |

Med Pandas kan du:
- Läsa in en sådan tabell från en fil (t.ex. Excel eller CSV)
- Titta på datan
- Filtrera (t.ex. visa bara elever med betyg > 80)
- Räkna medelvärden, summor, osv.
- Spara resultaten tillbaka till en fil

---

## Varför just Pandas?

Python i sig kan hantera listor och tabeller, men det är omständligt.  
Pandas är byggt exakt för tabelldata och gör allt mycket enklare och snabbare.

---

## Hur använder man Pandas?

Innan vi kan använda Pandas måste vi **importera** det.  
Att importera ett bibliotek är som att ta fram ett verktyg ur verktygslådan – du måste hämta det innan du kan använda det.

Konventionen (det alla gör) är att importera Pandas med smeknamnet **`pd`**  
– det är kortare att skriva och alla känner igen det.

```python
# 'import' = hämta biblioteket
# 'as pd' = ge det smeknamnet 'pd' så vi slipper skriva 'pandas' hela tiden
import pandas as pd

# Skriv ut versionen för att bekräfta att allt fungerar
print('Pandas-version:', pd.__version__)
```

---

## Pandas två viktigaste byggstenar

Pandas har två centrala datastrukturer som du kommer att använda hela tiden:

| Namn        | Vad det är                     | Analog i verkligheten |
|-------------|--------------------------------|----------------------|
| `Series`    | En enda kolumn med värden       | En kolumn i Excel    |
| `DataFrame` | En hel tabell (rader + kolumner)| Ett helt Excel-blad  |

Vi börjar med `DataFrame` direkt – det är den du kommer använda mest.

```python
# Vi skapar vår första DataFrame!
# pd.DataFrame() är funktionen som skapar en tabell.
# Vi skickar in ett dictionary (Python-sätt att spara nyckel-värde-par):
#   nyckel  = kolumnnamnet  (t.ex. 'namn')
#   värde   = en lista med data i den kolumnen

df = pd.DataFrame({
    'namn':  ['Anna', 'Bo', 'Carla'],
    'betyg': [85, 92, 78],
    'klass': ['9A', '9B', '9A'],
})

# Skriv ut tabellen – Jupyter visar den snyggt automatiskt
df
```

**Vad hände ovanför?**

1. Vi skapade tre kolumner: `namn`, `betyg`, `klass`
2. Varje kolumn har 3 värden — en per rad
3. Pandas lade automatiskt till ett **index** till vänster (0, 1, 2) — det är bara radnummer
4. Vi sparade tabellen i variabeln `df` (vanlig förkortning för DataFrame)

---

```python
# Vi kan kolla storleken på tabellen med .shape
# Resultatet är (antal rader, antal kolumner)
print('Storlek på tabellen:', df.shape)

# .shape[0] = antal rader
# .shape[1] = antal kolumner
print('Antal rader:', df.shape[0])
print('Antal kolumner:', df.shape[1])
```

```python
# Vi kan se vilka kolumnnamn tabellen har med .columns
print('Kolumnnamn:', df.columns.tolist())
```

```python
# Vi kan se vilken datatyp varje kolumn har med .dtypes
# object = text (kallas också 'string')
# int64  = heltal
# float64 = decimaltal
print(df.dtypes)
```
