```python
from pathlib import Path

print('Path ar laddad och redo.')
```

## Ovning 1 - Skapa Path-objekt

Varfor: Innan vi kan jobba med filer maste vi skapa adressen till filen eller mappen.

```python
project_root = Path('..')
raw_dir = Path('../data/raw')
processed_dir = Path('../data/processed')

print('Projektrot:', project_root.resolve())
print('Raw-dir:', raw_dir)
print('Processed-dir:', processed_dir)
```

## Ovning 2 - Skapa mappar sakert

Varfor: Kod som sparar filer far annars fel om mappen inte finns.

```python
practice_dir = Path('../data/processed/practice')
practice_dir.mkdir(parents=True, exist_ok=True)

print('Mappen finns nu:', practice_dir.exists())
print('Ar det en mapp:', practice_dir.is_dir())
```

## Ovning 3 - Bygg filvag med slash operator

Varfor: slash operatorn ar enklare och mindre felkanslig an att bygga textstrangar manuellt.

```python
notes_file = practice_dir / 'anteckningar.txt'
print('Filvag:', notes_file)
print('Filnamn:', notes_file.name)
print('Andelse:', notes_file.suffix)
print('Namn utan andelse:', notes_file.stem)
```

## Ovning 4 - Skriv text till fil

Varfor: Path kan skriva och lasa text direkt, utan extra bibliotek.

```python
text = 'Detta ar min forsta textfil skapad med Path.\nJag tranar pa pathlib.'
notes_file.write_text(text, encoding='utf-8')

print('Skrev fil:', notes_file.exists())
print('Storlek i byte:', notes_file.stat().st_size)
```

## Ovning 5 - Las text fran fil

Varfor: Du behover ofta kontrollera vad som sparats.

```python
innehall = notes_file.read_text(encoding='utf-8')
print('Innehall i filen:')
print(innehall)
```

## Ovning 6 - Lista filer i en mapp

Varfor: Bra nar du vill se vad som finns i en datamapp.

```python
print('Filer i practice-mappen:')
for p in practice_dir.iterdir():
    print('-', p.name)
```

## Ovning 7 - Hitta alla csv-filer

Varfor: I datajobb vill man ofta hitta alla filer av en viss typ.

```python
processed = Path('../data/processed')
csv_filer = list(processed.glob('*.csv'))

print('CSV-filer i data/processed:')
if csv_filer:
    for fil in csv_filer:
        print('-', fil.name)
else:
    print('Inga csv-filer hittades just nu.')
```
