# Lektion 4 – Filtrera och välja data

---

## Varför filtrera?

I verkligheten vill du sällan arbeta med HELA tabellen på en gång.  
Du vill t.ex.:

- Visa bara anställda i Stockholm
- Hitta alla med lön över 40 000
- Kombinera: Tech-anställda i Stockholm med hög lön

Pandas filtrering fungerar som att ställa frågor till tabellen.

```python
import pandas as pd
import numpy as np

# Samma data som förut
df = pd.DataFrame({
    'namn':      ['Anna', 'Bo', 'Carla', 'David', 'Eva', 'Frank', 'Greta', 'Hans'],
    'ålder':     [25, 31, 28, 33, 42, 35, 29, 38],
    'stad':      ['Stockholm', 'Göteborg', 'Stockholm', 'Malmö', 'Stockholm', 'Göteborg', 'Stockholm', 'Malmö'],
    'lön':       [32000, 45000, 38000, 41000, 55000, 39000, 36000, 48000],
    'avdelning': ['HR', 'Tech', 'HR', 'Tech', 'Tech', 'HR', 'Tech', 'HR'],
})

df
```

---
## Grundläggande filtrering – enkla villkor

Filtrering i Pandas fungerar i 2 steg:

1. Skapa ett **filter** (en True/False-kolumn) – vilka rader uppfyller villkoret?
2. Använd filtret på tabellen för att **välja ut** de raderna

```python
# STEG 1: Skapa ett filter
# df['stad'] == 'Stockholm' jämför varje värde i stad-kolumnen med 'Stockholm'
# Resultatet är True/False för varje rad
filter_stockholm = df['stad'] == 'Stockholm'

print('Filter (True = bor i Stockholm):')
print(filter_stockholm)
```

```python
# STEG 2: Använd filtret på tabellen
# df[filter] väljer bara de rader där filtret är True
stockholm_anstallda = df[filter_stockholm]

stockholm_anstallda
```

```python
# I praktiken skriver man oftast ihop det på en rad:
# df[ df['kolumn'] == värde ]
df[df['lön'] > 40000]
```

---
## Jämförelseoperatorer

Du kan använda alla dessa i dina filter:

| Operator | Betydelse | Exempel |
|---|---|---|
| `==` | Lika med | `df['stad'] == 'Malmö'` |
| `!=` | Inte lika med | `df['stad'] != 'Malmö'` |
| `>`  | Större än | `df['lön'] > 40000` |
| `<`  | Mindre än | `df['lön'] < 40000` |
| `>=` | Större än eller lika med | `df['ålder'] >= 30` |
| `<=` | Mindre än eller lika med | `df['ålder'] <= 30` |

```python
# Kombinera villkor med & (och) och | (eller)
# VIKTIGT: varje villkor MÅSTE ha egna parenteser!

# Hitta Tech-anställda MED hög lön (båda villkor måste vara True)
tech_och_hog_lon = df[(df['avdelning'] == 'Tech') & (df['lön'] > 40000)]

print('Tech-anställda med lön > 40 000:')
tech_och_hog_lon
```

```python
# Hitta de som bor i Stockholm ELLER Malmö (minst ett villkor behöver vara True)
stockholm_eller_malmo = df[(df['stad'] == 'Stockholm') | (df['stad'] == 'Malmö')]

print('Bor i Stockholm eller Malmö:')
stockholm_eller_malmo
```

```python
# .isin() är ett smidigare sätt att kolla om värdet finns i en lista
# Samma resultat som ovan men enklare att läsa
df[df['stad'].isin(['Stockholm', 'Malmö'])]
```

```python
# Filtrera på text som INNEHÅLLER en viss sträng
# .str.contains() söker efter texten var som helst i strängen
# case=False = ignorera stor/liten bokstav
df[df['namn'].str.contains('a', case=False)]
```

```python
# Negera ett filter med ~ (tilde = 'inte')
# Välj alla som INTE är i Tech
df[~(df['avdelning'] == 'Tech')]
```

---
## Filtrera och välja specifika kolumner på en gång

```python
# Visa bara namn och lön för Stockholm-anställda
# 1. Filtrera rader: df[df['stad'] == 'Stockholm']
# 2. Välj kolumner: [['namn', 'lön']]
df[df['stad'] == 'Stockholm'][['namn', 'lön']]
```

```python
# Räkna hur många rader som klarar ett villkor med .sum()
# True = 1, False = 0, så summan = antalet True-värden
antal_hog_lon = (df['lön'] > 40000).sum()
print(f'Antal anställda med lön över 40 000: {antal_hog_lon}')

# Eller som procent
andel_hog_lon = (df['lön'] > 40000).mean() * 100
print(f'Andel med lön över 40 000: {andel_hog_lon:.1f}%')
```

## Sammanfattning Lektion 4

| Vad vi lärde oss | Kod |
|---|---|
| Filtrera med ett villkor | `df[df['kol'] == värde]` |
| Filtrera: OCH | `df[(villkor1) & (villkor2)]` |
| Filtrera: ELLER | `df[(villkor1) \| (villkor2)]` |
| Filtrera: INTE | `df[~(villkor)]` |
| Kolla om värde finns i lista | `df['kol'].isin([...])` |
| Text innehåller | `df['kol'].str.contains('text')` |
| Räkna matchande rader | `(villkor).sum()` |
