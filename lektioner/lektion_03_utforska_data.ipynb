## Varför utforska data?

När du tar emot en ny datafil vet du ofta inte vad den innehåller.  
Steg 1 är alltid att **förstå datan** innan du börjar analysera den:

- Hur många rader och kolumner finns det?
- Vilka datatyper är det?
- Finns det saknade värden?
- Hur ser värdena ut i varje kolumn?

Pandas har inbyggda verktyg för precis detta.

```python
import pandas as pd
import numpy as np  # numpy behövs för att skapa NaN (saknade värden)

# Vi skapar en lite större och mer realistisk tabell att öva på
# NaN (Not a Number) = saknat värde, används i Pandas för 'okänt' / 'tomt'
df = pd.DataFrame({
    'namn':    ['Anna', 'Bo', 'Carla', 'David', 'Eva', 'Frank', 'Greta', 'Hans'],
    'ålder':   [25, 31, 28, np.nan, 42, 35, 29, 38],
    'stad':    ['Stockholm', 'Göteborg', 'Stockholm', 'Malmö', None, 'Göteborg', 'Stockholm', 'Malmö'],
    'lön':     [32000, 45000, 38000, 41000, 55000, 39000, np.nan, 48000],
    'avdelning': ['HR', 'Tech', 'HR', 'Tech', 'Tech', 'HR', 'Tech', 'HR'],
})

df
```

---
## Snabbkoll av tabellen

```python
# .head(n) visar de FÖRSTA n raderna (standard är 5 om du inte anger n)
# Användbart när du har en stor fil och bara vill se hur den ser ut
print('--- DE FÖRSTA 3 RADERNA ---')
df.head(3)
```

```python
# .tail(n) visar de SISTA n raderna
# Bra för att se om filen avslutades rätt
print('--- DE SISTA 3 RADERNA ---')
df.tail(3)
```

```python
# .info() ger en SAMMANFATTNING av hela tabellen
# Visar: antal rader, kolumnnamn, datatyp, hur många icke-tomma värden
# VIKTIGT: om 'Non-Null Count' är lägre än total antal rader = det finns saknade värden!
df.info()
```

**Läs av .info() ovanför:**

- `RangeIndex: 8 entries` → tabellen har 8 rader (0 till 7)
- `ålder`: 7 non-null → **1 rad har saknat värde** (David)
- `stad`: 7 non-null → **1 rad har saknat värde** (Eva)
- `lön`: 7 non-null → **1 rad har saknat värde** (Greta)
- `object` = text, `float64` = decimaltal (NaN tvingar Pandas att använda float)

```python
# .describe() ger statistik för NUMERISKA kolumner
# count  = antal icke-tomma värden
# mean   = medelvärde
# std    = standardavvikelse (hur spridda värdena är)
# min    = lägsta värde
# 25%    = 25:e percentilen (25% av värdena är lägre än detta)
# 50%    = medianvärdet
# 75%    = 75:e percentilen
# max    = högsta värde
df.describe()
```

```python
# .describe(include='all') inkluderar TEXTKOLUMNER också
# unique  = antal unika värden
# top     = det vanligaste värdet
# freq    = hur många gånger det vanligaste värdet förekommer
df.describe(include='all')
```

---
## Hitta saknade värden

```python
# .isna() returnerar True/False för varje cell – True = saknat värde
# .sum() räknar ihop True-värdena per kolumn (True = 1, False = 0)
# → Resultatet visar hur MÅNGA saknade värden varje kolumn har
print('Saknade värden per kolumn:')
print(df.isna().sum())
```

```python
# Visa procent saknade värden istället
# Vi dividerar med antal rader (len(df)) och multiplicerar med 100
procent_saknade = (df.isna().sum() / len(df)) * 100
print('Procent saknade värden per kolumn:')
print(procent_saknade.round(1))  # .round(1) = avrunda till 1 decimal
```

---
## Räkna unika värden och fördelningar

```python
# .value_counts() räknar hur många gånger varje unikt värde förekommer
# Perfekt för kategoriska kolumner (text som har ett begränsat antal alternativ)
print('Hur många bor i varje stad?')
print(df['stad'].value_counts())
```

```python
# .nunique() = antal unika värden i en kolumn
print('Unika städer:', df['stad'].nunique())
print('Unika avdelningar:', df['avdelning'].nunique())

# .unique() = lista på vilka unika värden som finns
print('Vilka städer:', df['stad'].unique())
```

```python
# .sort_values() sorterar tabellen efter en kolumn
# ascending=True  = stigande ordning (lägst värde först) — standard
# ascending=False = fallande ordning (högst värde först)
print('Sorterat efter lön, högt till lågt:')
df.sort_values('lön', ascending=False)
```
