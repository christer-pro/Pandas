## Del 1: Series – en enda kolumn med data

En **Series** är som en lista med ett index (radnummer/etiketter) vid sidan.  
Tänk på det som en **enda kolumn** i en tabell.

```
Index | Värde
  0   |  85
  1   |  92
  2   |  78
```

````python
import pandas as pd

# Skapa en enkel Series från en lista
# pd.Series() tar en lista och gör den till en kolumn
betyg = pd.Series([85, 92, 78, 65, 99])

print(betyg)
print()
print('Typ:', type(betyg))  # bekräftar att det är en Series
````

````python
# Vi kan ge Series ett eget index istället för 0, 1, 2...
# Här använder vi elevnamnen som index
betyg_namngivet = pd.Series(
    [85, 92, 78, 65, 99],
    index=['Anna', 'Bo', 'Carla', 'David', 'Eva']
)

print(betyg_namngivet)
print()

# Hämta ett specifikt värde med hjälp av indexnamnet
# Ungefär som att slå upp ett ord i ett lexikon
print('Bos betyg:', betyg_namngivet['Bo'])
````

````python
# En Series kan man räkna på precis som i matte
print('Medelvärde:', betyg_namngivet.mean())   # genomsnitt
print('Högsta betyg:', betyg_namngivet.max())   # max
print('Lägsta betyg:', betyg_namngivet.min())   # min
print('Summa:', betyg_namngivet.sum())          # summa
````

## Del 2: DataFrame – en hel tabell

En **DataFrame** är egentligen bara **flera Series sida vid sida**, en per kolumn.  
Det är den datastruktur du kommer använda 95% av tiden.

```
Index | namn  | betyg | klass
  0   | Anna  |  85   |  9A
  1   | Bo    |  92   |  9B
  2   | Carla |  78   |  9A
```

````python
# Skapa en DataFrame med mer data att öva på
df = pd.DataFrame({
    'namn':    ['Anna', 'Bo', 'Carla', 'David', 'Eva', 'Frank'],
    'betyg':   [85, 92, 78, 65, 99, 71],
    'klass':   ['9A', '9B', '9A', '9B', '9A', '9B'],
    'ålder':   [15, 16, 15, 16, 15, 16],
})

df
````

````python
# Välj en enda kolumn med hakparentes + kolumnnamn
# Detta ger oss tillbaka en Series!
betyg_kolumn = df['betyg']

print(betyg_kolumn)
print()
print('Typ:', type(betyg_kolumn))  # notera: det är en Series, inte en DataFrame
````

````python
# Välj FLERA kolumner med en lista inuti hakparentesen
# Notera de dubbla hakparenteserna: df[['kolumn1', 'kolumn2']]
# Inre []  = listan med kolumnnamn
# Yttre [] = 'hämta från DataFrame'
namn_och_betyg = df[['namn', 'betyg']]

namn_och_betyg
````

````python
# Välj en rad med .iloc (iloc = 'integer location' = radnummer)
# Räknar från 0, precis som i Python-listor
forsta_raden = df.iloc[0]  # rad nummer 0 = Anna

print(forsta_raden)
print()
print('Typ:', type(forsta_raden))  # en rad är också en Series!
````

````python
# Välj ett specifikt värde med .iloc[radnummer, kolumnnummer]
# rad 2, kolumn 1 = Carlas betyg
carlas_betyg = df.iloc[2, 1]
print('Carlas betyg:', carlas_betyg)

# Eller med kolumnnamn istället för nummer (.loc)
# .loc[radnummer, 'kolumnnamn']
carlas_betyg2 = df.loc[2, 'betyg']
print('Carlas betyg (loc):', carlas_betyg2)
````

````python
# Lägga till en ny kolumn är enkelt!
# Vi skapar en kolumn 'godkänd' som är True om betyg >= 70, annars False
# df['betyg'] >= 70 jämför VARJE betyg med 70 och ger True/False
df['godkänd'] = df['betyg'] >= 70

df
````

````python
# Ta bort en kolumn med .drop()
# axis=1 betyder 'ta bort en kolumn' (axis=0 skulle ta bort en rad)
# inplace=True betyder 'ändra df direkt, spara inte i ny variabel'
df.drop(columns=['godkänd'], inplace=True)

df
````

## Sammanfattning Lektion 2

| Vad vi lärde oss | Kod |
|---|---|
| Skapa en Series | `pd.Series([1, 2, 3])` |
| Beräkna medelvärde | `s.mean()` |
| Välj en kolumn | `df['kolumn']` |
| Välj flera kolumner | `df[['kol1', 'kol2']]` |
| Välj rad med nummer | `df.iloc[0]` |
| Välj värde med namn | `df.loc[2, 'kolumn']` |
| Lägga till kolumn | `df['ny'] = ...` |
| Ta bort kolumn | `df.drop(columns=['kolumn'])` |
